##### Creee una función que impute la media o la mediana, en función de la correlación entre una columna y el target. Su función debe considerar si la imputación es valida. Debe mostrar su funicón funcionando para tres datasets distintos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('train_HP.csv')
dataset.head()
```

```python
def CCA(df, x, z):
    mean_val = np.round(df.iloc[:,x].mean(), 0)
    print(f'la media es: {mean_val}')
    median_val = np.round(df.iloc[:,x].median(), 0)
    print(f'la mediana es: {median_val}')
    df_LF_meanImp = df.iloc[:,x].fillna(mean_val)
    df_LF_meadianImp = df.iloc[:,x].fillna(median_val)
    corr1 = np.corrcoef(df_LF_meanImp, df.iloc[:,z])[0,1]
    corr2 = np.corrcoef(df_LF_meadianImp, df.iloc[:,z])[0,1]
    print(corr1)
    print(corr2)
    if corr1 >= corr2:
        df.iloc[:,x] = df.iloc[:,x].fillna(mean_val)
    else:
        df.iloc[:,x] = df.iloc[:,x].fillna(median_val)
    print('Validacion Valor 0:')
    print(df.iloc[:,x].isnull().sum())
```

```python
CCA(dataset,3,80)
```
